```python
# Extract few features from the full info ['name','alternating','positive','alexander_polynomial']
import pandas as pd
D = pd.DataFrame(columns=['name','alternating','positive','alexander_polynomial'])
names = []
alternatings = []
positives = []
alexanders = []
for item in k[1:len(k)]:
    names.append(item['name'])
    alternatings.append(item['alternating'])
    positives.append(item['positive'])
    alexanders.append(item['alexander_polynomial'])
    print(f"{item['name']}: alternating = {item['alternating']}, alexander_polynomial = {item['alexander_polynomial']}, positive={item['positive']}")
```

```python
print(len(names), len(alternatings), len(positives), len(alexanders))
```

```python
D['name'] = names
D['alternating'] = alternatings
D['positive'] = positives
D['alexander_polynomial'] = alexanders
```
